```python
import pandas as pd
import numpy as np
import datetime
```

```python
file = "visualization/resources/url_df_updated.csv"

# read file into pandas
asmr = pd.read_csv(file)
asmr.head()
```

```python
asmrAnalysis = asmr[['channel', 'uploads', 'subs', 'total_views', 'channel_creation']]
asmrAnalysis
```

```python
asmrAnalysis_sort = asmrAnalysis.sort_values(by=['subs'], ascending=False)
```

```python
asmrAnalysis_sort
```

```python
asmrAnalysis_sort_by_views = asmrAnalysis.sort_values(by=['total_views'], ascending=False)
```

```python
topTen = asmrAnalysis_sort.head(10)
```

```python
topTen
```

```python
topTen['channel_creation_year'] = pd.DatetimeIndex(topTen['channel_creation']).year
```

```python
topTen
```

```python
topTenv2 = topTen[['channel', 'uploads', 'subs', 'total_views', 'channel_creation_year']]
```
