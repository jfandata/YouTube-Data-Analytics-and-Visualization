```python
import pandas as pd
import re
```

```python
asmr_channels = pd.read_csv("asmr.csv", encoding = "ISO-8859-1", header=None)
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
asmr_channels = asmr_channels.rename(columns={0: 'channel',  1: 'url'})

asmr_channels.head()
```

```python

asmr_channels["type"] = ["user" if 'user' in url else "channel" for url in asmr_channels['url']]
asmr_channels.head()
```

```python
user_df = asmr_channels[asmr_channels['url'].str.contains('user')]
channel_id_df = asmr_channels[asmr_channels['url'].str.contains('channel')]
```

```python
user_df
```

```python
user_list = user_df['url'].values.tolist()
channel_id_list = channel_id_df['url'].values.tolist()
```

```python
user_names = [re.sub(r'^(.*?)user/', "", user) for user in user_list]
channel_ids = [re.sub(r'^(.*?)channel/', "", user) for user in user_list]
```

```python
url = "https://socialblade.com/youtube/user/"
url2 = "https://socialblade.com/youtube/channel/"
```

```python
user_urls = [url + user + "/monthly" for user in user_names]
channel_id_urls = [url2 + chan + "/monthly" for chan in channel_ids]
```
