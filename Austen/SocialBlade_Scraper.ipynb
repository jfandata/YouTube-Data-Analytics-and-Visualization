```python
from bs4 import BeautifulSoup

import requests

import pandas as pd
```

```python
# Setup the BS4 scraper to get the Top 250 Youtube Channels in the Entertainment category

url = "socialblade.com/youtube/top/category/entertainment/mostsubscribed"

r  = requests.get("http://" +url)

data = r.text

soup = BeautifulSoup(data, "lxml")

```

```python


totalUploads = []

results = soup.find_all("span")

# Loop through results to retrieve total uploads
for result in results[10::4]:
    
    uploads = result.text
    
    totalUploads.append(uploads)
    
    
#     .find(style_="float: left; width: 50px; color:#888;")
#     style="color:#555;"
#     print(uploads)
```

```python
subscribers = []

for result in results[11::4]:
    
    subs = result.text
    
    subscribers.append(subs)
#     print(subs)
```

```python
totalVideoViews = []

for result in results[12::4]:
    
    videoViews = result.text
    
    totalVideoViews.append(videoViews)
#     print(videoViews)
```

```python
channelName = []

results = soup.find_all("a")

for result in results[157:407]:
    
    channel = result.text
    
    channelName.append(channel)
#     print(channel)
```

```python
# Print the length of each list to ensure they line up correctly

totalVideoViews = totalVideoViews[0:250]
subscribers = subscribers[0:250]
totalUploads = totalUploads[0:250]

print(len(channelName))
print(len(totalVideoViews))
print(len(subscribers))
print(len(totalUploads))
```

```python
# Create the dataframe 

youtube_entertainment_df = pd.DataFrame({'Username': channelName, 'Total Uploads': totalUploads,
                           'Subscribers': subscribers, 'Total Video Views': totalVideoViews})

youtube_entertainment_df.head(15)
```
